```python
import matplotlib.pyplot as plt
import numpy as np

points = []

path = "./input.txt"

with open(path) as f:
    lines = f.readlines()
    for line in lines:
        tokens = line.split(",")
        x = int(tokens[0])
        y = int(tokens[1])
        z = int(tokens[2])
        points.append((x,y,z))


x, y, z = np.indices((21, 21, 21))

voxelarray = (x == -1) & (y == -1) & (z == -1)
for (xd, yd, zd) in points:
    voxelarray = voxelarray | (x == xd) & (y == yd) & (z == zd)

ax = plt.figure().add_subplot(projection='3d')
ax.voxels(voxelarray, facecolors='red', edgecolor='k')

plt.show()
```
